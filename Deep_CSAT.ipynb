## ***Hypothesis Testing***

### **Hypothetical Statement - 1**

**Higher response time significantly reduces customer satisfaction (CSAT).**

#### **1. State Your research hypothesis as a null hypothesis and alternate hypothesis.**

```
H₀ (Null Hypothesis) : There is no significant relationship between response time and CSAT score.

H₁ (Alternative Hypothesis) : There is a significant negative relationship between response time and CSAT score.
```

#### **2. Perform an appropriate statistical test.**

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import spearmanr

# Drop missing values
temp = df[['response_time_minutes', 'CSAT Score']].dropna()

# Spearman correlation
corr, p_value = spearmanr(temp['response_time_minutes'],
                          temp['CSAT Score'])

print("Spearman Correlation:", corr)
print("P-value:", p_value)
````

Since p-value < 0.05 and correlation is negative, we conclude that higher response time significantly reduces customer satisfaction.

There is a statistically significant but weak negative relationship between response time and customer satisfaction. As response time increases, CSAT slightly decreases.

### **Hypothetical Statement - 2**

**Customer satisfaction differs significantly across support channels.**

#### **1. State Your research hypothesis as a null hypothesis and alternate hypothesis.**

```
H₀ (Null Hypothesis) : Mean CSAT score is the same across all channels.

H₁ (Alternative Hypothesis) : At least one channel has significantly different mean CSAT.
```

#### **2. Perform an appropriate statistical test.**

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import kruskal

group1 = df[df['channel_name'] == 'Inbound']['CSAT Score']
group2 = df[df['channel_name'] == 'Outcall']['CSAT Score']
group3 = df[df['channel_name'] == 'Email']['CSAT Score']

stat, p_value = kruskal(group1, group2, group3)

print("Kruskal-Wallis Statistic:", stat)
print("P-value:", p_value)
````

Since p-value < 0.05, customer satisfaction significantly varies across support channels.

There is a statistically significant difference in customer satisfaction across support channels (p < 0.001). However, given the large sample size, the practical effect may be small.

### **Hypothetical Statement - 3**

**Agent shift timing significantly affects customer satisfaction.**

#### **1. State Your research hypothesis as a null hypothesis and alternate hypothesis.**

```
H₀ (Null Hypothesis) : Mean CSAT score is equal across all shifts.

H₁ (Alternative Hypothesis) : At least one shift has significantly different mean CSAT.
```

#### **2. Perform an appropriate statistical test.**

````python
# Perform Statistical Test to obtain P-

shifts = df['Agent Shift'].unique()

groups = [df[df['Agent Shift'] == shift]['CSAT Score']
          for shift in shifts]

stat, p_value = kruskal(*groups)

print("Kruskal-Wallis Statistic:", stat)
print("P-value:", p_value)
````

There is a statistically significant difference in customer satisfaction across Agent Shift (p < 0.001). However, given the large sample size, the practical effect may be small.

## **DL Model Implementation**

**Predict CSAT Score using 'Customer Remarks' feature only.**

````python
# import some libraries for Deep Learing.

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU, Bidirectional, Dropout, Embedding, BatchNormalization
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam, AdamW, RMSprop, SGD
````

````python
# create a dataset which contain only "Customer Remarks" and 'CSAT Score'.

df_remark = df[['Customer Remarks', 'CSAT Score']].dropna()
len(df_remark)
````

"df_remark" contains only those data points which is reviewed by customer.

````python
# Clean Text

import re

def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z]', '', text)
  return text

df_remark['Customer Remarks'] = df_remark['Customer Remarks'].apply(clean_text)
````

This function will takes only alphbetical characters and remove punctuations, special characters and numbers.

````python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_words = 10000
max_len = 50

# Tokenization.

tokenizer = Tokenizer(num_words = max_words, oov_token = '<OOV>')
tokenizer.fit_on_texts(df_remark['Customer Remarks'])

sequences = tokenizer.texts_to_sequences(df_remark['Customer Remarks'])

# Padding.

X = pad_sequences(sequences, maxlen=max_len, padding='post')
y = df_remark['CSAT Score'] - 1

# spliting.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
````

**Tokenization :** Tokenization will tokenize text into words and convert the words into numbers.

**Padding :** Padding will take 0 in the end of sentece if that sentece's lengthh is less than max_len value to maintain vector length.

Then splited the data by 80-20.

````python
# Model Training and Evaluation.

model = Sequential([
    Embedding(input_dim = max_words, output_dim = 64, input_length = max_len),
    Bidirectional(SimpleRNN(64, activation = 'sigmoid', return_sequences = True)),
    # Dropout(0.2),
    # # BatchNormalization(),
    Bidirectional(SimpleRNN(32, activation = 'sigmoid', return_sequences = True)),
    # # Dropout(0.2),
    # BatchNormalization(),
    Bidirectional(SimpleRNN(16, activation='sigmoid')),
    Dense(5, activation = 'softmax')
])

model.compile(
    optimizer = AdamW(learning_rate = 4e-5, weight_decay = 4e-5),
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

history = model.fit(
    X_train, y_train,
    epochs = 10,
    validation_split = 0.2, batch_size = 16
)

loss, accuracy = model.evaluate(X_test, y_test)

    # Plot
plt.figure(figsize = (12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy')
plt.legend(['Train', 'Validation'])

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.legend(['Train', 'Validation'])

plt.show()
````
